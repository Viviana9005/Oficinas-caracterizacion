```python

import pandas as pd

import matplotlib.pyplot as plt
```

```python
file = pd.read_excel('Base_Venta_Oficinas.xlsx')
df = pd.DataFrame(file)
df
```

```python
descriptivas=df.describe()
descriptivas
```

# Estadísticas por zona

```python
x=df.groupby(['Zona'])[['Precio Venta']].count()
x
```

```python
plot = x.plot.pie(y='Precio Venta', figsize=(10, 10), legend=None)
plt.savefig("Numero_Oficinas_Zona.jpg", bbox_inches='tight')
```

```python
precioventa_zona=df.groupby(['Zona'])[['Precio Venta']].median()
precioventa_zona
```

```python
valoradmin_zona=df.groupby(['Zona'])[['Valor Administración']].median()
valoradmin_zona
```

```python
Estrato_zona=df.groupby(['Zona'])[['Estrato']].median()
Estrato_zona
```

```python
par_zona=df.groupby(['Zona'])[['No. Parqueaderos']].median()
par_zona
```

```python
par_zona.plot(kind='bar',color='purple',legend=None, figsize=(10,10))
plt.savefig("Zona_parqueaderos.jpg", bbox_inches='tight', legend=None)
```

```python
m2_zona=df.groupby(['Zona'])[['M2']].median()
m2_zona
```

```python
pisos_zona=df.groupby(['Zona'])[['No. Pisos']].median()
pisos_zona
```

```python
baños_zona=df.groupby(['Zona'])[['No. Baños']].median()
baños_zona
```
